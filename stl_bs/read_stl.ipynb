```python
from stl import mesh
from mpl_toolkits import mplot3d
from matplotlib import pyplot
import stltovoxel
```

```python
# Create a new plot
figure = pyplot.figure()
axes = figure.add_subplot(111, projection='3d')


# Load the STL files and add the vectors to the plot
your_mesh = mesh.Mesh.from_file('model3.stl')
collection = mplot3d.art3d.Poly3DCollection(your_mesh.vectors, linewidths=1, alpha=1)
collection.set_facecolor([0.5, 0.5, 1])
collection.set_edgecolor('k')

axes.add_collection3d(collection)

# Auto scale to the mesh size
scale = your_mesh.points.flatten()
axes.auto_scale_xyz(scale, scale, scale)

# Show the plot to the screen
pyplot.show()
```

```python
stltovoxel.convert_file('model3.stl', 'output.xyz', resolution=100, voxel_size=0.25)
```
